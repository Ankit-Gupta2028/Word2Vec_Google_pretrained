## NLTK = Natural Language Toolkit ✅
It is a Python library used for NLP (Natural Language Processing).

It helps you work with text data easily.

✅ What NLTK is used for?

With NLTK you can do:

✅ Tokenization (split words/sentences),
✅ Stopwords removal (remove “is, the, a…”),
✅ Stemming (playing → play),
✅ Lemmatization (better than stemming),
✅ POS tagging (noun/verb/adjective),
✅ Text classification,
✅ N-grams (2-word, 3-word combos)

```python
import nltk

nltk.download("punkt", download_dir="/opt/anaconda3/envs/annstream/nltk_data")
nltk.download("punkt_tab", download_dir="/opt/anaconda3/envs/annstream/nltk_data")
nltk.data.path.append("/opt/anaconda3/envs/annstream/nltk_data")
from nltk.tokenize import sent_tokenize
sent_tokenize(corpus)
```

```python
## Tokenization
#convert paragraph in --->>> sentence
from nltk.tokenize import sent_tokenize
documents=sent_tokenize(corpus)
```

```python
## Tokenization
#convert paragraph in --->>> words
from nltk.tokenize import word_tokenize
words=word_tokenize(corpus)
words
```

```python
for sentences in documents:
    print(word_tokenize(sentences))
```

```python
from nltk.tokenize import wordpunct_tokenize
# wordpunct_tokenize() is an NLTK tokenizer ✅ that splits a sentence into words AND punctuation separately.
```

```python
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenize=TreebankWordTokenizer()
tokenize.tokenize(corpus)
```
