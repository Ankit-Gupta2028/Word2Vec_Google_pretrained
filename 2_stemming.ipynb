### Porter Stemmer


```python
from nltk.stem import PorterStemmer
stemming=PorterStemmer()
words=['eating','eats','eaten','writes','writing','programming','programs','history','finally','finalized']
for word in words:
    print(word+"---->"+stemming.stem(word))
```

```python
stemming.stem('congratulations')
```

## RegexStemmer
RegexStemmer in NLTK is a type of stemmer âœ… that uses Regular Expressions (regex) to remove endings (suffixes) from words.

So basically:

âœ… RegexStemmer = custom stemming using regex rules

### âœ… Why use RegexStemmer?

Because sometimes you want your own stemming rules instead of using Porter/Snowball.

Example:
You want to remove only:

ing

ed

s

Then RegexStemmer is perfect âœ…

```python
from nltk.stem import RegexpStemmer
```

```python
reg_Stemmer=RegexpStemmer('ing$|s$|e$|able$',min=4)
reg_Stemmer.stem('ingeating')
```

## Snowball Stemmer
Snowball Stemmer is an NLP stemmer in NLTK âœ… used to reduce words to their root form (stem).

It is an improved version of Porter Stemmer and is also called:
âœ… Why Snowball Stemmer is popular?

âœ… Faster
âœ… More accurate than PorterStemmer
âœ… Supports many languages like:

English

French

German

Spanish

Italian etc.

```python
from nltk.stem import SnowballStemmer
Stem_SnowballStemmer=SnowballStemmer("english")

for word in words:
    print(word+"----->>"+Stem_SnowballStemmer.stem(word))
```

```python
## Difference blt SnowballStemmer and PorterStemmer

#PorterStemmer
print("PorterStemmer--> "+stemming.stem("fairly"))

#SnowballStemmer
print("SnowballStemmer--> "+Stem_SnowballStemmer.stem("fairly"))
```
