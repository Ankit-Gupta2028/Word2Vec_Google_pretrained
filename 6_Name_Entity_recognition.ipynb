## Named Entity Recognition (NER)
Named Entity Recognition (NER) is an NLP (Natural Language Processing) task where we find ‚Äúreal-world names‚Äù in text and label their type.

‚úÖ It answers: ‚ÄúWho/What is being talked about?‚Äù

‚úÖ What NER detects (Entity Types)

NER finds important entities like:

PERSON ‚Üí people names
‚úÖ ‚ÄúVirat Kohli‚Äù

ORG (Organization) ‚Üí company, college, group
‚úÖ ‚ÄúGoogle‚Äù, ‚ÄúIIT Delhi‚Äù

GPE / LOCATION ‚Üí country, city, place
‚úÖ ‚ÄúIndia‚Äù, ‚ÄúMumbai‚Äù

DATE / TIME ‚Üí dates and time
‚úÖ ‚Äú21 January 2026‚Äù, ‚Äú6 PM‚Äù

MONEY ‚Üí currency amounts
‚úÖ ‚Äú‚Çπ5000‚Äù, ‚Äú$100‚Äù

PRODUCT ‚Üí product names
‚úÖ ‚ÄúiPhone 15‚Äù

‚úÖ Example

Sentence:
üëâ ‚ÄúAnkit Gupta lives in India and works at Microsoft.‚Äù

NER Output:

Ankit Gupta ‚Üí PERSON

India ‚Üí GPE (Location)

Microsoft ‚Üí ORG

```python
sentence="The Eiffel Tower was built from 1887 to 1889 by Gustave Eiffel, whose company specialized in building metal frameworks and structures."
```

```python
import nltk

# Download all required NLTK resources (run once)
nltk.download("punkt")
nltk.download("averaged_perceptron_tagger")
nltk.download("maxent_ne_chunker_tab")
nltk.download("words")
```

```python
tokens = nltk.word_tokenize(sentence)
tokens
```

```python
Pos_tag=nltk.pos_tag(tokens)
Pos_tag
```

```python
nltk.ne_chunk(Pos_tag)
```
