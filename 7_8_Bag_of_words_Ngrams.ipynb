```python
import pandas as pd
messages=pd.read_csv('SMSSpamCollection.csv',sep='\t',names=["label","massage"])
messages.head()
```

```python
## Data preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-Z]',' ',messages['massage'][i]) #Apart from A To Z any thing there replace with blank
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)


```

```python
corpus
```

```python
# Create Bag of Words model
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=100,binary=True) # it means out of all word take 2500 words which have max frequency
# Binary equal to true means Bag of words act as binary Bag of words.
```

```python
X=cv.fit_transform(corpus).toarray()
X
```

```python
cv.vocabulary_
```
